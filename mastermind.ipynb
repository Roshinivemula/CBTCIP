```python
import random

def get_valid_guess(length):
    while True:
        guess = input(f"Player, Enter your {length}-digit guess: ")
        if guess.isdigit() and len(guess) == length:
            return guess
        else:
            print(f"Invalid input. Please enter a {length}-digit number.")

def provide_feedback(secret, guess):
    correct_positions = sum(1 for s, g in zip(secret, guess) if s == g)
    correct_digits = sum(min(secret.count(digit), guess.count(digit)) for digit in set(guess))
    return correct_positions, correct_digits

def play_round(secret_setter, guesser, secret_length):
    secret = input(f"{secret_setter}, please set your {secret_length}-digit secret number: ")
    while not (secret.isdigit() and len(secret) == secret_length):
        print(f"Invalid input. Please enter a {secret_length}-digit number.")
        secret = input(f"{secret_setter}, please set your {secret_length}-digit secret number: ")
    
    print("\n" * 50)  # Clear the screen for the guesser

    tries = 0
    while True:
        guess = get_valid_guess(secret_length)
        tries += 1
        if guess == secret:
            print(f"{guesser} guessed the number {secret} in {tries} tries!")
            return tries
        correct_positions, correct_digits = provide_feedback(secret, guess)
        print(f"Correct digits in the correct place: {correct_positions}")
        print(f"Correct digits but in the wrong place: {correct_digits - correct_positions}")

def main():
    secret_length = int(input("Enter the length of the secret number: "))

    print("\nPlayer 1's turn to set the number.")
    p1_tries = play_round("Player 1", "Player 2", secret_length)

    print("\nPlayer 2's turn to set the number.")
    p2_tries = play_round("Player 2", "Player 1", secret_length)

    if p1_tries > p2_tries:
        print("\nPlayer 1 wins and is crowned Mastermind!")
    elif p2_tries > p1_tries:
        print("\nPlayer 2 wins and is crowned Mastermind!")
    else:
        print("\nIt's a tie! Both players are Masterminds!")

if __name__ == "__main__":
    main()
```
